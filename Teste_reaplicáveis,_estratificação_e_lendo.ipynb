```python
import pandas as pd #biblioteca Pandas foi importada e agora pode ser referenciada usando o alias pd3....................................................................................................................................6666666666666666/

uri = "https://gist.githubusercontent.com/guilhermesilveira/b9dd8e4b62b9e22ebcb9c8e89c271de4/raw/c69ec4b708fba03c445397b6a361db4345c83d7a/tracking.csv"
dados = pd.read_csv(uri) #le o arquivo csv (localiza o csv)
dados.head() #mostra a "cabeça" do csv
```

```python
#armazenando no y de quem comprou os ingressos
y=dados["comprou"]
y.head() #mostra a "cabeça" do y
```

```python
#armazenando no x as colunas que a maquina vai fazer a previsao
x=dados[["inicial", "palestras",	"contato", "patrocinio"]]
x.head() #mostra a "cabeça" do x
```

```python
#para saber quantas linhas e colunas tem nos dados
dados.shape
```

```python
#separando os itens para treino e teste
# O treino e teste eles tem que ser diferente! Se eu passar o treino com os
# mesmos dados do meu teste, a máquina decora, assim como nos decoramos. Fica quase como "over fit"
treino_x = x[:75]#meu treino vai ser as primeiras 75 linhas
treino_y = y[:75]

teste_x = x[75:]#meu teste vai a partir da linha 75
teste_y = y[75:]

print(f"Treinaremos com {len(treino_x)} itens")
print(f"Testaremos com {len(teste_y)} itens")
```

```python
#vamos treinar o modelo
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

modelo = LinearSVC()
#aqui ele esta treinando
# treinando o modelo a receber os dados treino_x para que saiam com a saida em treino_y
modelo.fit(treino_x, treino_y)

previsoes = modelo.predict(teste_x)


acuracia = accuracy_score(teste_y, previsoes) * 100 #o primeiro algumento é resultado depois as previcoes
print(f"A acurácia foi de {acuracia:.2f}%")

#over fit é quando o modelo fica 100% perfeito
```

Usando a biblioteca para separar treino e teste


```python
from sklearn.model_selection import train_test_split

#essa função ela automaticamente separada os dados em 4 partes! (treino x e y, teste x e y)
treino_x, teste_x, treino_y, teste_y = train_test_split(x,y)

print(f"Treinaremos com {len(treino_x)} itens")
print(f"Testaremos com {len(teste_y)} itens")
```
